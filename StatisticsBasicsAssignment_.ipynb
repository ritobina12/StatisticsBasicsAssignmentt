**Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

Answer: Descriptive statistics summarize and describe the main features of a dataset, while inferential statistics make predictions or inferences about a population based on sample data.

Examples:
•	Descriptive Statistics: Calculating the average test score of a class, creating a histogram of student grades, or reporting the standard deviation of ages in a survey.
•	Inferential Statistics: Using a sample of voters to predict election results, testing whether a new drug is more effective than a placebo based on clinical trial data, or estimating the average height of all adults in a country from a sample.
Key differences:
1.	Descriptive statistics deal with summarizing existing data, while inferential statistics make conclusions beyond the immediate data.
2.	Descriptive statistics don't involve probability, while inferential statistics rely on probability theory.
3.	Descriptive statistics are about "what is" while inferential statistics are about "what might be."

**Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**

Answer: Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population. It's essential when studying the entire population is impractical due to constraints of time, cost, or accessibility.
Random Sampling (Simple Random Sampling):
•	Every member of the population has an equal chance of being selected
•	No subgroups or strata are considered
•	Example: Drawing names from a hat containing all employee names to select survey participants
•	Advantages: Simple to implement, each member has equal selection probability
•	Disadvantages: May underrepresent small subgroups in heterogeneous populations
Stratified Sampling:
•	The population is divided into homogeneous subgroups called strata
•	Random samples are taken from each stratum
•	Example: Dividing a school population by grade level and then randomly sampling from each grade
•	Advantages: Ensures representation of all subgroups, can improve precision
•	Disadvantages: Requires knowledge of population structure, more complex to implement
Key differences:
1.	Stratified sampling divides the population into groups first, while random sampling doesn't
2.	Stratified sampling ensures all subgroups are represented, while random sampling might miss small groups
3.	Stratified sampling often provides more precise estimates for heterogeneous populations

**Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**

Answer:
Definitions:
•	Mean: The arithmetic average, calculated by summing all values and dividing by the number of values. Formula: Mean = (Σx)/n
•	Median: The middle value when data is ordered from smallest to largest (or the average of the two middle values for even-numbered datasets).
•	Mode: The most frequently occurring value in a dataset (a dataset can have multiple modes or no mode).
Importance of Measures of Central Tendency:
1.	Data Summary: They provide a single value that represents the center of the data distribution.
2.	Comparison: Allow easy comparison between different datasets or groups.
3.	Decision Making: Help in making informed decisions based on typical values.
4.	Foundation for Further Analysis: Serve as the basis for more complex statistical measures and tests.
5.	Understanding Distribution: Together, they help understand the shape of the data distribution.
Example: For the dataset [3, 5, 7, 7, 9]:
•	Mean = (3+5+7+7+9)/5 = 6.2
•	Median = 7 (middle value)
•	Mode = 7 (most frequent)
Each measure has strengths:
•	Mean uses all data but is affected by outliers
•	Median is robust to outliers but doesn't use all information
•	Mode shows most common value but may not represent center

**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

Answer:
Skewness measures the asymmetry of the probability distribution of a real-valued random variable about its mean.
•	Positive skew: The tail on the right side is longer or fatter (mean > median)
•	Negative skew: The tail on the left side is longer or fatter (mean < median)
•	Zero skew: Symmetrical distribution (mean = median)
Kurtosis measures the "tailedness" of the probability distribution (peakedness and tail heaviness).
•	Leptokurtic: High kurtosis (more peaked, heavier tails)
•	Mesokurtic: Medium kurtosis (normal distribution)
•	Platykurtic: Low kurtosis (less peaked, lighter tails)
Implications of Positive Skew:
1.	The majority of data points are concentrated on the left side of the distribution
2.	There are unusually high values pulling the mean up above the median
3.	The tail extends toward higher values
4.	Common in data where values have a natural minimum but no fixed maximum (e.g., income, house prices)
5.	May require data transformation for certain statistical analyses
Example: Income data typically shows positive skew - most people earn moderate amounts, while a few earn much more, creating a long right tail.




**Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**


```python
from statistics import mean, median, mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate measures
mean_value = mean(numbers)
median_value = median(numbers)
mode_value = mode(numbers)

# Display results
print("Numbers:", numbers)
print("\nMean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

**Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60]**

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate covariance
cov_matrix = np.cov(list_x, list_y)
covariance = cov_matrix[0, 1]

# Calculate correlation
corr_matrix = np.corrcoef(list_x, list_y)
correlation = corr_matrix[0, 1]

# Display results
print("Dataset X:", list_x)
print("Dataset Y:", list_y)
print("\nCovariance:", covariance)
print("Correlation coefficient:", correlation)
```

**Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**

```python
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=False, patch_artist=True)
plt.title("Boxplot of Data Distribution")
plt.xlabel("Values")
plt.grid(True)
plt.show()

# Calculate quartiles and IQR for explanation
import numpy as np
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Q1 (25th percentile): {q1}")
print(f"Q3 (75th percentile): {q3}")
print(f"IQR: {iqr}")
print(f"Lower bound (Q1 - 1.5*IQR): {lower_bound}")
print(f"Upper bound (Q3 + 1.5*IQR): {upper_bound}")
print(f"Outliers: {outliers}")
```


Interpretation:
1.	The box represents the interquartile range (IQR) from Q1 (17.25) to Q3 (23.25)
2.	The line inside the box is the median (~21.5)
3.	The whiskers extend to the lowest (12) and highest (29) values within 1.5*IQR of the quartiles
4.	The value 35 is marked as an outlier (beyond upper bound of 32.25)
5.	The median is slightly above the center of the box, indicating a mild right skew
6.	The longer upper whisker compared to lower also suggests right skewness
7.	The outlier (35) is substantially higher than the rest of the data points


**Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
•	Explain how you would use covariance and correlation to explore this relationship.
•	Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]**

Answer:

Explanation: To explore the relationship between advertising spend and daily sales:
1.	Covariance would tell us the direction of the relationship:
o	Positive covariance: As spending increases, sales tend to increase
o	Negative covariance: As spending increases, sales tend to decrease
o	Zero covariance: No linear relationship However, covariance is hard to interpret as its magnitude depends on the units of measurement.
2.	Correlation (Pearson's r) would give us a standardized measure:
o	Ranges from -1 to 1
o	1: Perfect positive linear relationship
o	-1: Perfect negative linear relationship
o	0: No linear relationship
o	Values between 0.7-1.0 or -0.7 to -1.0 typically indicate strong relationships
We would:
1.	First compute both measures
2.	Examine the correlation coefficient for strength and direction
3.	Consider creating a scatter plot for visual confirmation
4.	Interpret the results in business context

```python
#Python Implementation:
import numpy as np
import matplotlib.pyplot as plt

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Calculate correlation
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(advertising_spend, daily_sales, color='blue', s=100)
plt.title("Advertising Spend vs Daily Sales")
plt.xlabel("Advertising Spend ($)")
plt.ylabel("Daily Sales ($)")
plt.grid(True)

# Add regression line
m, b = np.polyfit(advertising_spend, daily_sales, 1)
plt.plot(advertising_spend, m*np.array(advertising_spend) + b, color='red')

plt.show()

print("Correlation coefficient:", correlation)
```

Interpretation:
•	The correlation coefficient of 0.99 indicates an extremely strong positive linear relationship
•	This means increases in advertising spend are strongly associated with increases in daily sales
•	The scatter plot with regression line would show points almost perfectly aligned along the line
•	The marketing team can be confident that their advertising spend has a strong association with sales
•	However, correlation doesn't imply causation - other factors may be involved


**Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
•	Explain which summary statistics and visualizations you'd use.
•	Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**

Answer:

Analysis Approach:
1.	Summary Statistics:
o	Central Tendency:
	Mean: Average satisfaction level
	Median: Middle score (less affected by extremes)
	Mode: Most common score(s)
o	Dispersion:
	Range: Min and max scores
	Standard deviation: Measure of spread
	Quartiles: 25th, 50th (median), 75th percentiles
o	Shape:
	Skewness: Asymmetry of distribution
	Kurtosis: Peakedness
2.	Visualizations:
o	Histogram: Shows frequency distribution of scores
o	Boxplot: Displays median, quartiles, and outliers
o	Bar chart: For categorical representation of counts
o	Pie chart: Could show proportion of satisfied vs. unsatisfied


```python
#Python Implementation:
import matplotlib.pyplot as plt
from statistics import mean, median, mode, stdev
import numpy as np

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate summary statistics
stats = {
    "Mean": mean(survey_scores),
    "Median": median(survey_scores),
    "Mode": mode(survey_scores),
    "Standard Deviation": stdev(survey_scores),
    "Minimum": min(survey_scores),
    "Maximum": max(survey_scores),
    "Range": max(survey_scores) - min(survey_scores)
}

print("Summary Statistics:")
for key, value in stats.items():
    print(f"{key}: {value:.2f}" if isinstance(value, float) else f"{key}: {value}")

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(survey_scores, bins=range(4, 12), edgecolor='black', align='left')
plt.title("Customer Satisfaction Scores Distribution")
plt.xlabel("Satisfaction Score (1-10)")
plt.ylabel("Frequency")
plt.xticks(range(4, 11))
plt.grid(axis='y', alpha=0.5)

# Add mean and median lines
plt.axvline(stats["Mean"], color='red', linestyle='dashed', linewidth=1, label=f'Mean: {stats["Mean"]:.2f}')
plt.axvline(stats["Median"], color='green', linestyle='dashed', linewidth=1, label=f'Median: {stats["Median"]}')
plt.legend()

plt.show()
```

Interpretation:
1.	The histogram shows:
o	Most common scores between 7-8
o	Few responses at extremes (4 and 10)
o	Approximately bell-shaped with slight right skew
2.	Summary statistics:
o	Mean (7.27) and median (7) are close, suggesting reasonable symmetry
o	Mode is 7 (most frequent score)
o	Standard deviation of 1.62 indicates moderate spread
o	Range of 6 shows variability in responses
3.	Business implications:
o	Majority of customers are satisfied (scores 7+)
o	Some room for improvement (scores ≤6)
o	The distribution suggests the product is generally well-received
o	Could segment analysis to understand low scores
